### Bowling pins problem statement
https://www.hackerrank.com/challenges/bowling-pins/problem

```python
import os
import sys

win = [0]
for n in range(1,301) :
    # maximum number of pins
    s = set()
    for a in range(n//2+1) :
        # maximum number of chances for a particular player.
        s.add(win[a]^win[n-1-a])
        # for each chance a player can take down one or two pins 
        # [n-1-a = Total number of pins - one pin -no of chances ] 
        # win[a]^win[n-1-a] - this gives the grundy number for each chance
        s.add(win[a]^win[n-2-a])
        # similarly in this line of code we are taking down two pins at each chance
        # now a set will create for each N pins and ...
        # the set values are based on the grundy number between chances and remaining no of pins

    m = 0
    while m in s :
        m += 1
    win.append(m)
    # this will find the minimum excludant value of the set ( win =Mex)

def solve(S) :
    r = 0
    for s in S.split('X') :
        # we need to split the pins based on where it is knocked down by using X
        r ^= win[len(s)]
        #this will find the XOR(mex) values between each split 
        #if the XOR(mex)=0 it denotes we are losing otherwise we are winning
    if r:
        return "WIN" 
    else:
        return "LOSE"

T = int(input())
for _ in range(T) :
    N = int(input())
    S = input()
    print(solve(S))
```

### This problem approached on the basis of Sprague grundy theorem, by using grundy number we can tell whether a person is winning (or) losing a position.

We need to find XOR ( set of possibel positions )
- if XOR=0 ( losing )
- if XOR!=0 ( winning )

In this game we can knock down any single pin, or any two adjacent pins at one throw of a bowling ball.

Grundy number defines in what state the game is going on.

Mex ( Minimum excludant value ) of set of numbers is the smallest non-negative number not present in the set.

suppose our set value is = [ 0,1,3 ] here mex value is 2.
