```python
title_basics_movie = title_basics1[(title_basics1["titleType"] == "movie")
            | (title_basics1["titleType"] == "tvMovie")]
title_basics_movie["titleType"] = "movie"
```

```python
title_basics_movie.head()
```

```python
Series = title_basics1[(title_basics1["titleType"] == "tvSeries") | (title_basics1["titleType"] == "tvMiniSeries")]
Series["titleType"] = "series"
```

```python
Episode = title_basics1[(title_basics1["titleType"]== "tvEpisode")]
Episode['titleType'] = 'episode'
```

```python
title_basicsNew = pd.concat([title_basics_movie, Series,Episode], ignore_index=True)
title_basicsNew.head()
```
