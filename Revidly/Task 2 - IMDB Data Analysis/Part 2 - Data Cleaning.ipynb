# Name_basics

```python
name_basics5.head()
```

```python
name_basics5["primaryProfession"].replace(np.nan, "None", inplace = True)
name_basics = pd.DataFrame(name_basics5.primaryProfession.str.split(',')
                                         .tolist(), index=name_basics5.nconst).stack()
name_basics = name_basics.reset_index()[['nconst', 0]] 
name_basics.rename(columns = {0: 'primaryProfession'}, inplace=True)
```

```python
name_basics.head()
```

```python
none = name_basics5[name_basics5["primaryProfession"] != "None"]
```

```python
namebasics_final = pd.merge(none, name_basics, left_on=['nconst'], right_on = ['nconst'])
namebasics_final.drop(columns = ["primaryProfession_x"], axis = True, inplace=True)
namebasics_final.rename(columns = {"primaryProfession_y": 'primaryProfession'}, inplace=True)
```

```python
namebasics_final.head()
```

```python
namebasics_final_actor_actress = namebasics_final[(namebasics_final["primaryProfession"] == "actor") | 
                                    (namebasics_final["primaryProfession"] == "actress")]
namebasics_final_director = namebasics_final[namebasics_final["primaryProfession"] == "director"]
```

```python
namebasics_final_actor_actress.head()
```

```python
namebasics_final_actor_actress.shape
```

```python
namebasics_final_director.head()
```

```python
namebasics_final_director.shape
```

```python
name_final = pd.concat([namebasics_final_actor_actress, namebasics_final_director], ignore_index=True)
name_final.head()
```

```python
name_final.shape
```
